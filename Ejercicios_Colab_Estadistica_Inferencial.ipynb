## Ejercicio 1: Intervalo de Confianza para la Media Poblacional

Se desea estimar la altura promedio de estudiantes universitarios. En una muestra de 50 estudiantes, se obtuvo una media de 170 cm. Se sabe que la desviación estándar poblacional es de 10 cm. Construye un intervalo de confianza del 95% para la media poblacional.

### 1. Definir las hipótesis nula y alternativa.

Aunque para la construcción de un intervalo de confianza no se definen formalmente hipótesis nula y alternativa como en una prueba de hipótesis, si quisiéramos plantear una prueba relacionada con la media poblacional ($\mu$), podríamos considerar:

*   **Hipótesis Nula ($H_0$):** La altura promedio de los estudiantes universitarios es igual a un valor específico ($\mu = \mu_0$).
*   **Hipótesis Alternativa ($H_1$):** La altura promedio de los estudiantes universitarios es diferente a un valor específico ($\mu \neq \mu_0$), o mayor ($\mu > \mu_0$), o menor ($\mu < \mu_0$).

En este caso, nos enfocaremos en la construcción del intervalo de confianza.

### 2. Establecer el nivel de significancia α.

Para un intervalo de confianza del 95%, el nivel de significancia ($\alpha$) es:

$\alpha = 1 - \text{Nivel de Confianza} = 1 - 0.95 = 0.05$

### 3. Identificar el estadístico de prueba (Z, t, χ²) y su distribución.

Dado que conocemos la desviación estándar poblacional ($\sigma$) y la muestra es relativamente grande ($n=50 > 30$), utilizamos el **estadístico Z** y la **distribución normal estándar**.

### 4. Determinar el valor crítico y la región de rechazo.

Para un intervalo de confianza bilateral del 95%, necesitamos encontrar el valor Z crítico que deja $\alpha/2 = 0.05/2 = 0.025$ en cada cola de la distribución normal estándar.

El valor crítico $Z_{\alpha/2}$ para un 95% de confianza es 1.96.

La región de rechazo estaría en las colas de la distribución, más allá de los valores críticos si estuviéramos realizando una prueba de hipótesis. Para el intervalo de confianza, usamos el valor crítico para definir los límites del intervalo.

### 5. Calcular el estadístico de prueba y el valor p.

En la construcción de un intervalo de confianza, no calculamos un estadístico de prueba en el mismo sentido que en una prueba de hipótesis. En su lugar, utilizamos la fórmula del intervalo de confianza para la media poblacional cuando se conoce la desviación estándar poblacional:

Intervalo de Confianza = $\bar{x} \pm Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}$

Donde:
*   $\bar{x}$ es la media muestral (170 cm)
*   $Z_{\alpha/2}$ es el valor crítico de Z para el nivel de confianza (1.96 para 95%)
*   $\sigma$ es la desviación estándar poblacional (10 cm)
*   $n$ es el tamaño de la muestra (50)

El valor p se relaciona con la probabilidad de obtener un resultado tan extremo o más extremo que el observado, asumiendo que la hipótesis nula es verdadera. En el contexto de un intervalo de confianza, el valor p no se calcula directamente de la misma manera que en una prueba de hipótesis. El intervalo mismo nos da un rango de valores plausibles para el parámetro poblacional.

A continuación, calcularemos el intervalo de confianza usando Python.

```
import numpy as np
from scipy.stats import norm

# Datos proporcionados
media_muestral = 170
desviacion_estandar_poblacional = 10
tamaño_muestra = 50
nivel_confianza = 0.95

# Nivel de significancia
alpha = 1 - nivel_confianza

# Valor crítico de Z para un intervalo bilateral
valor_critico_z = norm.ppf(1 - alpha/2)

# Error estándar
error_estandar = desviacion_estandar_poblacional / np.sqrt(tamaño_muestra)

# Margen de error
margen_error = valor_critico_z * error_estandar

# Calcular el intervalo de confianza
limite_inferior = media_muestral - margen_error
limite_superior = media_muestral + margen_error

print(f"Media muestral: {media_muestral}")
print(f"Desviación estándar poblacional: {desviacion_estandar_poblacional}")
print(f"Tamaño de la muestra: {tamaño_muestra}")
print(f"Nivel de confianza: {nivel_confianza}")
print(f"Valor crítico de Z: {valor_critico_z:.2f}")
print(f"Error estándar: {error_estandar:.2f}")
print(f"Margen de error: {margen_error:.2f}")
print(f"Intervalo de Confianza del {nivel_confianza*100}%: [{limite_inferior:.2f}, {limite_superior:.2f}]")
```

## Ejercicio 11: Tamaño de Muestra para Proporción Poblacional (Vacunación)

Para determinar el tamaño de muestra necesario para estimar una proporción poblacional, utilizamos la siguiente fórmula cuando se tiene una estimación previa de la proporción ($\hat{p}$):

$n = \frac{Z_{\alpha/2}^2 \cdot \hat{p}(1 - \hat{p})}{E^2}$

Donde:
*   $n$ es el tamaño de muestra necesario.
*   $Z_{\alpha/2}$ es el valor crítico de Z para el nivel de confianza.

$n = \frac{Z_{\alpha/2}^2 \cdot 0.25}{E^2}$

Para determinar el tamaño de muestra necesario para estimar una proporción poblacional, utilizamos la fórmula:

$n = \frac{Z_{\alpha/2}^2 \cdot \hat{p}(1 - \hat{p})}{E^2}$

## Ejercicio 21: Prueba de Hipótesis para la Media Poblacional (Peso de Cereal)

Una empresa afirma que sus paquetes de cereal tienen un peso promedio de 500 g. Se sospecha que la máquina está descalibrada y subllenando. Se toma una muestra de 40 paquetes, obteniendo una media de 495 g. La desviación estándar poblacional es σ = 15 g. Con un nivel de de significancia α =0,05.

### 1. Plantee las hipótesis nula y alternativa.

*   **Hipótesis Nula ($H_0$):** El peso promedio de los paquetes es de 500 g ($\mu = 500$).
*   **Hipótesis Alternativa ($H_1$):** El peso promedio de los paquetes es menor a 500 g ($\mu < 500$). (Subllenando)

### 2. Establezca el nivel de significancia α.

El nivel de significancia proporcionado es $\alpha = 0.05$.

### 3. Identifique el estadístico de prueba (Z, t, χ²) y su distribución.

Dado que conocemos la desviación estándar poblacional ($\sigma = 15$) y el tamaño de la muestra es grande ($n=40 > 30$), utilizamos el **estadístico Z** y la **distribución normal estándar**.

### 4. Determine el valor crítico y la región de rechazo.

Para una prueba de hipótesis unilateral izquierda con $\alpha = 0.05$, necesitamos encontrar el valor Z crítico que deja 0.05 en la cola izquierda de la distribución normal estándar.

El valor crítico $Z_{\alpha}$ para una prueba unilateral izquierda con $\alpha = 0.05$ es aproximadamente -1.645.

La región de rechazo es $Z < -1.645$.

### 5. Calcule el estadístico de prueba y el valor p.

Calcularemos el estadístico de prueba Z utilizando la fórmula:

$Z = \frac{\bar{x} - \mu_0}{\sigma / \sqrt{n}}$

Donde:
*   $\bar{x}$ es la media muestral (495 g)
*   $\mu_0$ es la media poblacional bajo la hipótesis nula (500 g)
*   $\sigma$ es la desviación estándar poblacional (15 g)
*   $n$ es el tamaño de la muestra (40)

Calcularemos el valor p, que es la probabilidad de obtener un estadístico de prueba tan extremo o más extremo que el observado, asumiendo que la hipótesis nula es verdadera. Para una prueba unilateral izquierda, el valor p es $P(Z \le \text{estadístico de prueba})$.

A continuación, calcularemos el estadístico de prueba Z y el valor p usando Python.

```
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns

# Datos proporcionados (Ejercicio 21)
media_muestral_ej21 = 495
media_poblacional_nula_ej21 = 500
desviacion_estandar_poblacional_ej21 = 15
tamaño_muestra_ej21 = 40
alpha_ej21 = 0.05

# Calcular el estadístico de prueba Z
error_estandar_ej21 = desviacion_estandar_poblacional_ej21 / np.sqrt(tamaño_muestra_ej21)
estadistico_z_ej21 = (media_muestral_ej21 - media_poblacional_nula_ej21) / error_estandar_ej21

# Calcular el valor p (prueba unilateral izquierda)
valor_p_ej21 = norm.cdf(estadistico_z_ej21)

# Determinar el valor crítico de Z (prueba unilateral izquierda)
valor_critico_z_ej21 = norm.ppf(alpha_ej21)


print(f"--- Ejercicio 21 ---")
print(f"Estadístico de prueba Z: {estadistico_z_ej21:.3f}")
print(f"Valor p: {valor_p_ej21:.4f}")
print(f"Valor crítico de Z: {valor_critico_z_ej21:.3f}")

# --- Paso 6: Graficar la distribución ---
plt.figure(figsize=(10, 6))
x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x), label='Distribución Normal Estándar')

# Región de rechazo (unilateral izquierda)
x_rechazo = np.linspace(-4, valor_critico_z_ej21, 100)
plt.fill_between(x_rechazo, norm.pdf(x_rechazo), color='red', alpha=0.5, label='Región de Rechazo (α=0.05)')

# Valor crítico
plt.axvline(valor_critico_z_ej21, color='red', linestyle='--', label=f'Valor Crítico Z = {valor_critico_z_ej21:.3f}')

# Estadístico calculado
plt.axvline(estadistico_z_ej21, color='blue', linestyle='-', label=f'Estadístico Z = {estadistico_z_ej21:.3f}')

plt.title('Distribución Normal Estándar con Región de Rechazo, Valor Crítico y Estadístico Z')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

Dado que el tamaño de la muestra es pequeño ($n=15 < 30$) y conocemos la desviación estándar muestral ($s = 4$), utilizamos el **estadístico t** y la **distribución t de Student**.

### 4. Determine los grados de libertad y el valor crítico, además de la región crítica y la región de rechazo.

Los grados de libertad son $n-1 = 15-1 = 14$.

El valor crítico $t_{\alpha, n-1}$ para una prueba unilateral derecha con $\alpha = 0.05$ y 14 grados de libertad es aproximadamente 1.761.

$t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}$

*   $\bar{x}$ es la media muestral (28 minutos)
*   $\mu_0$ es la media poblacional bajo la hipótesis nula (25 minutos)

## Ejercicio 27: Prueba de Hipótesis para la Media Poblacional (Calificación de Matemáticas)

Un colegio afirma que el promedio de su clase en matemáticas es 80/100. Una muestra de 12 estudiantes tiene una media de 75 y s = 8. Con un nivel de significancia de α = 0,01.

### 1. Plantee las hipótesis nula y alternativa.

*   **Hipótesis Nula ($H_0$):** El promedio de la clase en matemáticas es 80 ($\mu = 80$).
*   **Hipótesis Alternativa ($H_1$):** El promedio de la clase en matemáticas es diferente a 80 ($\mu \neq 80$). (Prueba bilateral, ya que la afirmación es sobre un valor específico y se quiere ver si es diferente)

### 2. Establezca el nivel de significancia α.

El nivel de significancia proporcionado es $\alpha = 0.01$.

### 3. Identifique el estadístico de prueba (Z, t, χ²) y su distribución.

Dado que el tamaño de la muestra es pequeño ($n=12 < 30$) y conocemos la desviación estándar muestral ($s = 8$), utilizamos el **estadístico t** y la **distribución t de Student**.

### 4. Determine los grados de libertad y el valor crítico, además de la región crítica y la región de rechazo.

Los grados de libertad son $n-1 = 12-1 = 11$.

Para una prueba de hipótesis bilateral con $\alpha = 0.01$ y 11 grados de libertad, necesitamos encontrar los valores t críticos que dejan $\alpha/2 = 0.01/2 = 0.005$ en cada cola de la distribución t de Student.

Los valores críticos $t_{\alpha/2, n-1}$ para una prueba bilateral con $\alpha = 0.01$ y 11 grados de libertad son aproximadamente -3.106 y +3.106.

La región de rechazo es $t < -3.106$ o $t > +3.106$.

### 5. Calcule el estadístico de prueba y el valor p.

Calcularemos el estadístico de prueba t utilizando la fórmula:

$t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}$

Donde:
*   $\bar{x}$ es la media muestral (75)
*   $\mu_0$ es la media poblacional bajo la hipótesis nula (80)
*   $s$ es la desviación estándar muestral (8)
*   $n$ es el tamaño de la muestra (12)

Calcularemos el valor p. Para una prueba bilateral, el valor p es el doble de la probabilidad en la cola más allá del estadístico de prueba. Es decir, $p = 2 \cdot P(t > |\text{estadístico de prueba}|)$ con $n-1$ grados de libertad.

A continuación, calcularemos el estadístico de prueba t y el valor p usando Python.

```
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt
import seaborn as sns

# Datos proporcionados (Ejercicio 27)
media_muestral_ej27 = 75
media_poblacional_nula_ej27 = 80
desviacion_estandar_muestral_ej27 = 8
tamaño_muestra_ej27 = 12
alpha_ej27 = 0.01
grados_libertad_ej27 = tamaño_muestra_ej27 - 1

# Calcular el estadístico de prueba t
error_estandar_ej27 = desviacion_estandar_muestral_ej27 / np.sqrt(tamaño_muestra_ej27)
estadistico_t_ej27 = (media_muestral_ej27 - media_poblacional_nula_ej27) / error_estandar_ej27

# Calcular el valor p (prueba bilateral)
valor_p_ej27 = 2 * (1 - t.cdf(abs(estadistico_t_ej27), grados_libertad_ej27))

# Determinar los valores críticos de t (prueba bilateral)
valor_critico_t_inferior_ej27 = t.ppf(alpha_ej27/2, grados_libertad_ej27)
valor_critico_t_superior_ej27 = t.ppf(1 - alpha_ej27/2, grados_libertad_ej27)

print(f"--- Ejercicio 27 ---")
print(f"Estadístico de prueba t: {estadistico_t_ej27:.3f}")
print(f"Valor p: {valor_p_ej27:.4f}")
print(f"Grados de libertad: {grados_libertad_ej27}")
print(f"Valores críticos de t: [{valor_critico_t_inferior_ej27:.3f}, {valor_critico_t_superior_ej27:.3f}]")

# --- Paso 6: Graficar la distribución ---
plt.figure(figsize=(10, 6))
x = np.linspace(-4, 4, 1000)
plt.plot(x, t.pdf(x, grados_libertad_ej27), label=f'Distribución t (GL={grados_libertad_ej27})')

# Región de rechazo (bilateral)
x_rechazo_inferior = np.linspace(-4, valor_critico_t_inferior_ej27, 100)
plt.fill_between(x_rechazo_inferior, t.pdf(x_rechazo_inferior, grados_libertad_ej27), color='red', alpha=0.5, label='Región de Rechazo (α/2=0.005)')

x_rechazo_superior = np.linspace(valor_critico_t_superior_ej27, 4, 100)
plt.fill_between(x_rechazo_superior, t.pdf(x_rechazo_superior, grados_libertad_ej27), color='red', alpha=0.5) # No repetir etiqueta

# Valores críticos
plt.axvline(valor_critico_t_inferior_ej27, color='red', linestyle='--', label=f'Valores Críticos t = {valor_critico_t_inferior_ej27:.3f}, {valor_critico_t_superior_ej27:.3f}')
plt.axvline(valor_critico_t_superior_ej27, color='red', linestyle='--')

# Estadístico calculado
plt.axvline(estadistico_t_ej27, color='blue', linestyle='-', label=f'Estadístico t = {estadistico_t_ej27:.3f}')

plt.title(f'Distribución t (GL={grados_libertad_ej27}) con Región de Rechazo, Valores Críticos y Estadístico t')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

Los grados de libertad son $n-1 = 20-1 = 19$.

Para una prueba de hipótesis bilateral con $\alpha = 0.05$ y 19 grados de libertad, necesitamos encontrar los valores t críticos que dejan $\alpha/2 = 0.05/2 = 0.025$ en cada cola de la distribución t de Student.

Los valores críticos $t_{\alpha/2, n-1}$ para una prueba bilateral con $\alpha = 0.05$ y 19 grados de libertad son aproximadamente -2.093 y +2.093.

La región de rechazo es $t < -2.093$ o $t > +2.093$.

### 5. Calcule el estadístico de prueba y el valor p.

Calcularemos el estadístico de prueba t utilizando la fórmula:

$t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}$

Donde:
*   $\bar{x}$ es la media muestral (9,800 horas)
*   $\mu_0$ es la media poblacional bajo la hipótesis nula (10,000 horas)
*   $s$ es la desviación estándar muestral (300 horas)
*   $n$ es el tamaño de la muestra (20)

Calcularemos el valor p. Para una prueba bilateral, el valor p es el doble de la probabilidad en la cola más allá del estadístico de prueba. Es decir, $p = 2 \cdot P(t > |\text{estadístico de prueba}|)$ con $n-1$ grados de libertad.

A continuación, calcularemos el estadístico de prueba t y el valor p usando Python.

```
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt
import seaborn as sns

# Datos proporcionados (Ejercicio 30)
media_muestral_ej30 = 9800
media_poblacional_nula_ej30 = 10000
desviacion_estandar_muestral_ej30 = 300
tamaño_muestra_ej30 = 20
alpha_ej30 = 0.05
grados_libertad_ej30 = tamaño_muestra_ej30 - 1

# Calcular el estadístico de prueba t
error_estandar_ej30 = desviacion_estandar_muestral_ej30 / np.sqrt(tamaño_muestra_ej30)
estadistico_t_ej30 = (media_muestral_ej30 - media_poblacional_nula_ej30) / error_estandar_ej30

# Calcular el valor p (prueba bilateral)
valor_p_ej30 = 2 * (1 - t.cdf(abs(estadistico_t_ej30), grados_libertad_ej30))

# Determinar los valores críticos de t (prueba bilateral)
valor_critico_t_inferior_ej30 = t.ppf(alpha_ej30/2, grados_libertad_ej30)
valor_critico_t_superior_ej30 = t.ppf(1 - alpha_ej30/2, grados_libertad_ej30)

print(f"--- Ejercicio 30 ---")
print(f"Estadístico de prueba t: {estadistico_t_ej30:.3f}")
print(f"Valor p: {valor_p_ej30:.4f}")
print(f"Grados de libertad: {grados_libertad_ej30}")
print(f"Valores críticos de t: [{valor_critico_t_inferior_ej30:.3f}, {valor_critico_t_superior_ej30:.3f}]")

# --- Paso 6: Graficar la distribución ---
plt.figure(figsize=(10, 6))
x = np.linspace(-4, 4, 1000)
plt.plot(x, t.pdf(x, grados_libertad_ej30), label=f'Distribución t (GL={grados_libertad_ej30})')

# Región de rechazo (bilateral)
x_rechazo_inferior = np.linspace(-4, valor_critico_t_inferior_ej30, 100)
plt.fill_between(x_rechazo_inferior, t.pdf(x_rechazo_inferior, grados_libertad_ej30), color='red', alpha=0.5, label='Región de Rechazo (α/2=0.025)')

x_rechazo_superior = np.linspace(valor_critico_t_superior_ej30, 4, 100)
plt.fill_between(x_rechazo_superior, t.pdf(x_rechazo_superior, grados_libertad_ej30), color='red', alpha=0.5) # No repetir etiqueta

# Valores críticos
plt.axvline(valor_critico_t_inferior_ej30, color='red', linestyle='--', label=f'Valores Críticos t = {valor_critico_t_inferior_ej30:.3f}, {valor_critico_t_superior_ej30:.3f}')
plt.axvline(valor_critico_t_superior_ej30, color='red', linestyle='--')

# Estadístico calculado
plt.axvline(estadistico_t_ej30, color='blue', linestyle='-', label=f'Estadístico t = {estadistico_t_ej30:.3f}')

plt.title(f'Distribución t (GL={grados_libertad_ej30}) con Región de Rechazo, Valores Críticos y Estadístico t')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

## Ejercicio 31: Prueba de Hipótesis para la Varianza Poblacional (Tiempo de Llamadas)

Una compañía de telefonía afirma que la varianza del tiempo de duración de las llamadas es de σ² = 2,5 minutos². Se toma una muestra de n = 16 llamadas y se obtiene una varianza muestral de s² = 4. ¿Hay evidencia al 5% de significancia de que la varianza poblacional difiere de 2.5?

Para probar una hipótesis sobre la varianza poblacional ($\sigma^2$), utilizamos el estadístico de prueba Chi-cuadrado ($\chi^2$). La fórmula es:

$\chi^2 = \frac{(n - 1) s^2}{\sigma_0^2}$

Donde:
*   $n$ es el tamaño de la muestra.
*   $s^2$ es la varianza muestral.
*   $\sigma_0^2$ es la varianza poblacional bajo la hipótesis nula.

La distribución del estadístico de prueba es la distribución Chi-cuadrado con $n-1$ grados de libertad.

Siguiendo la estructura solicitada:

### 1. Plantee las hipótesis nula y alternativa.

*   **Hipótesis Nula ($H_0$):** La varianza poblacional del tiempo de duración de las llamadas es 2.5 minutos² ($\sigma^2 = 2.5$).
*   **Hipótesis Alternativa ($H_1$):** La varianza poblacional del tiempo de duración de las llamadas es diferente a 2.5 minutos² ($\sigma^2 \neq 2.5$). (Prueba bilateral)

### 2. Establezca el nivel de significancia α.

El nivel de significancia proporcionado es $\alpha = 0.05$.

### 3. Identifique el estadístico de prueba (Z, t, χ²) y su distribución.

Dado que estamos probando una hipótesis sobre la varianza poblacional, utilizamos el **estadístico Chi-cuadrado ($\chi^2$)** y la **distribución Chi-cuadrado**.

### 4. Determine los grados de libertad y el valor crítico, además de la región crítica y la región de rechazo.

Los grados de libertad son $n-1 = 16-1 = 15$.

Para una prueba de hipótesis bilateral con $\alpha = 0.05$ y 15 grados de libertad, necesitamos encontrar dos valores críticos de Chi-cuadrado: $\chi^2_{1-\alpha/2, n-1}$ y $\chi^2_{\alpha/2, n-1}$.

*   Valor crítico inferior: $\chi^2_{0.975, 15}$
*   Valor crítico superior: $\chi^2_{0.025, 15}$

Los valores críticos de Chi-cuadrado para una prueba bilateral con $\alpha = 0.05$ y 15 grados de libertad son aproximadamente 6.262 y 27.488.

La región de rechazo es $\chi^2 < 6.262$ o $\chi^2 > 27.488$.

### 5. Calcule el estadístico de prueba y el valor p.

Calcularemos el estadístico de prueba Chi-cuadrado utilizando la fórmula:

$\chi^2 = \frac{(n - 1) s^2}{\sigma_0^2}$

Donde:
*   $n$ es el tamaño de la muestra (16).
*   $s^2$ es la varianza muestral (4).
*   $\sigma_0^2$ es la varianza poblacional bajo la hipótesis nula (2.5).

Calcularemos el valor p. Para una prueba bilateral, el valor p es el doble de la probabilidad en la cola más allá del estadístico de prueba (hacia la cola más cercana). Es decir, si el estadístico $\chi^2$ es menor que la mediana, $p = 2 \cdot P(\chi^2 \le \text{estadístico }\chi^2)$. Si es mayor, $p = 2 \cdot P(\chi^2 \ge \text{estadístico }\chi^2)$.

A continuación, calcularemos el estadístico de prueba Chi-cuadrado y el valor p usando Python.

```
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt
import seaborn as sns

# Datos proporcionados (Ejercicio 31)
varianza_muestral_ej31 = 4
varianza_poblacional_nula_ej31 = 2.5
tamaño_muestra_ej31 = 16
alpha_ej31 = 0.05
grados_libertad_ej31 = tamaño_muestra_ej31 - 1

# Calcular el estadístico de prueba Chi-cuadrado
estadistico_chi2_ej31 = (grados_libertad_ej31 * varianza_muestral_ej31) / varianza_poblacional_nula_ej31

# Determinar los valores críticos de Chi-cuadrado (prueba bilateral)
valor_critico_chi2_inferior_ej31 = chi2.ppf(alpha_ej31/2, grados_libertad_ej31)
valor_critico_chi2_superior_ej31 = chi2.ppf(1 - alpha_ej31/2, grados_libertad_ej31)

# Calcular el valor p (prueba bilateral)
# Si el estadístico está a la izquierda de la mediana (GL/2), calculamos 2*CDF. Si está a la derecha, 2*(1-CDF).
# La mediana de una distribución chi2 con GL > 2 es aproximadamente (GL * (1 - 2/(9*GL))^3). Para GL=15, la mediana es ~14.3
if estadistico_chi2_ej31 < grados_libertad_ej31: # Aproximación simple para saber en qué cola está
     valor_p_ej31 = 2 * chi2.cdf(estadistico_chi2_ej31, grados_libertad_ej31)
else:
     valor_p_ej31 = 2 * (1 - chi2.cdf(estadistico_chi2_ej31, grados_libertad_ej31))


print(f"--- Ejercicio 31 ---")
print(f"Estadístico de prueba Chi-cuadrado: {estadistico_chi2_ej31:.3f}")
print(f"Valor p: {valor_p_ej31:.4f}")
print(f"Grados de libertad: {grados_libertad_ej31}")
print(f"Valores críticos de Chi-cuadrado: [{valor_critico_chi2_inferior_ej31:.3f}, {valor_critico_chi2_superior_ej31:.3f}]")

# --- Paso 6: Graficar la distribución ---
plt.figure(figsize=(10, 6))
x = np.linspace(0, 40, 1000) # Rango adecuado para GL=15
plt.plot(x, chi2.pdf(x, grados_libertad_ej31), label=f'Distribución $\chi^2$ (GL={grados_libertad_ej31})')

# Región de rechazo (bilateral)
x_rechazo_inferior = np.linspace(0, valor_critico_chi2_inferior_ej31, 100)
plt.fill_between(x_rechazo_inferior, chi2.pdf(x_rechazo_inferior, grados_libertad_ej31), color='red', alpha=0.5, label='Región de Rechazo (α/2=0.025)')

x_rechazo_superior = np.linspace(valor_critico_chi2_superior_ej31, 40, 100)
plt.fill_between(x_rechazo_superior, chi2.pdf(x_rechazo_superior, grados_libertad_ej31), color='red', alpha=0.5) # No repetir etiqueta

# Valores críticos
plt.axvline(valor_critico_chi2_inferior_ej31, color='red', linestyle='--', label=f'Valores Críticos $\chi^2$ = {valor_critico_chi2_inferior_ej31:.3f}, {valor_critico_chi2_superior_ej31:.3f}')
plt.axvline(valor_critico_chi2_superior_ej31, color='red', linestyle='--')

# Estadístico calculado
plt.axvline(estadistico_chi2_ej31, color='blue', linestyle='-', label=f'Estadístico $\chi^2$ = {estadistico_chi2_ej31:.3f}')

plt.title(f'Distribución $\chi^2$ (GL={grados_libertad_ej31}) con Región de Rechazo, Valores Críticos y Estadístico $\chi^2$')
plt.xlabel('Valor $\chi^2$')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

*   **Hipótesis Nula ($H_0$):** La varianza poblacional en el grosor es 0.5 mm² ($\sigma^2 = 0.5$). (Se asume el límite inferior de la afirmación del control de calidad para la hipótesis nula).
*   **Hipótesis Alternativa ($H_1$):** La varianza poblacional en el grosor es menor a 0.5 mm² ($\sigma^2 < 0.5$). (Prueba unilateral izquierda)

### 2. Establezca el nivel de significancia α.

El nivel de significancia proporcionado es $\alpha = 0.05$.

### 3. Identifique el estadístico de prueba (Z, t, χ²) y su distribución.

Dado que estamos probando una hipótesis sobre la varianza poblacional, utilizamos el **estadístico Chi-cuadrado ($\chi^2$)** y la **distribución Chi-cuadrado**.

### 4. Determine los grados de libertad y el valor crítico, además de la región crítica y la región de rechazo.

Los grados de libertad son $n-1 = 12-1 = 11$.

Para una prueba de hipótesis unilateral izquierda con $\alpha = 0.05$ y 11 grados de libertad, necesitamos encontrar el valor crítico de Chi-cuadrado que deja 0.05 en la cola izquierda de la distribución Chi-cuadrado.

El valor crítico inferior: $\chi^2_{\alpha, n-1} = \chi^2_{0.05, 11}$

El valor crítico de Chi-cuadrado para una prueba unilateral izquierda con $\alpha = 0.05$ y 11 grados de libertad es aproximadamente 4.575.

La región de rechazo es $\chi^2 < 4.575$.

### 5. Calcule el estadístico de prueba y el valor p.

Calcularemos el estadístico de prueba Chi-cuadrado utilizando la fórmula:

$\chi^2 = \frac{(n - 1) s^2}{\sigma_0^2}$

Donde:
*   $n$ es el tamaño de la muestra (12).
*   $s^2$ es la varianza muestral (0.3).
*   $\sigma_0^2$ es la varianza poblacional bajo la hipótesis nula (0.5).

Calcularemos el valor p. Para una prueba unilateral izquierda, el valor p es la probabilidad de obtener un estadístico de prueba tan extremo o más extremo (menor) que el observado, asumiendo que la hipótesis nula es verdadera. Es decir, $p = P(\chi^2 \le \text{estadístico }\chi^2)$.

A continuación, calcularemos el estadístico de prueba Chi-cuadrado y el valor p usando Python.

```
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt
import seaborn as sns

# Datos proporcionados (Ejercicio 33)
varianza_muestral_ej33 = 0.3
varianza_poblacional_nula_ej33 = 0.5
tamaño_muestra_ej33 = 12
alpha_ej33 = 0.05
grados_libertad_ej33 = tamaño_muestra_ej33 - 1

# Calcular el estadístico de prueba Chi-cuadrado
estadistico_chi2_ej33 = (grados_libertad_ej33 * varianza_muestral_ej33) / varianza_poblacional_nula_ej33

# Determinar el valor crítico de Chi-cuadrado (prueba unilateral izquierda)
valor_critico_chi2_inferior_ej33 = chi2.ppf(alpha_ej33, grados_libertad_ej33)

# Calcular el valor p (prueba unilateral izquierda)
valor_p_ej33 = chi2.cdf(estadistico_chi2_ej33, grados_libertad_ej33)

print(f"--- Ejercicio 33 ---")
print(f"Estadístico de prueba Chi-cuadrado: {estadistico_chi2_ej33:.3f}")
print(f"Valor p: {valor_p_ej33:.4f}")
print(f"Grados de libertad: {grados_libertad_ej33}")
print(f"Valor crítico de Chi-cuadrado: {valor_critico_chi2_inferior_ej33:.3f}")

# --- Paso 6: Graficar la distribución ---
plt.figure(figsize=(10, 6))
x = np.linspace(0, 30, 1000) # Rango adecuado para GL=11
plt.plot(x, chi2.pdf(x, grados_libertad_ej33), label=f'Distribución $\chi^2$ (GL={grados_libertad_ej33})')

# Región de rechazo (unilateral izquierda)
x_rechazo = np.linspace(0, valor_critico_chi2_inferior_ej33, 100)
plt.fill_between(x_rechazo, chi2.pdf(x_rechazo, grados_libertad_ej33), color='red', alpha=0.5, label='Región de Rechazo (α=0.05)')

# Valor crítico
plt.axvline(valor_critico_chi2_inferior_ej33, color='red', linestyle='--', label=f'Valor Crítico $\chi^2$ = {valor_critico_chi2_inferior_ej33:.3f}')

# Estadístico calculado
plt.axvline(estadistico_chi2_ej33, color='blue', linestyle='-', label=f'Estadístico $\chi^2$ = {estadistico_chi2_ej33:.3f}')

plt.title(f'Distribución $\chi^2$ (GL={grados_libertad_ej33}) con Región de Rechazo, Valor Crítico y Estadístico $\chi^2$')
plt.xlabel('Valor $\chi^2$')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

Comparamos el valor p (0.1244) con el nivel de significancia ($\alpha = 0.05$). Dado que el valor p (0.1244) es mayor que $\alpha$ (0.05), no rechazamos la hipótesis nula ($H_0$).
